### Playing The Videos

```python
# Playing the fist video 
video_1 = cv.VideoCapture("video1.mp4")
while True :
    _, frame = video_1.read()
    if not _:
        break
    cv.imshow("This is the my first video", frame)
    if cv.waitKey(20) & 0xFF == ord("q"):
        break

video_1.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv

# Open the video file
video_1 = cv.VideoCapture("video1.mp4")

# Loop to read frames from the video
while True:
    # Read a single frame
    _, frame = video_1.read()

    # If the frame was not read successfully, the video has ended
    if not _:
        break

    # Display the frame in a window
    cv.imshow("This is my first video", frame)

    # Wait for a key press for 20 milliseconds
    # If the 'q' key is pressed, break the loop
    if cv.waitKey(20) & 0xFF == ord("q"):
        break

# Release the video capture object
video_1.release()

# Close all OpenCV windows
cv.destroyAllWindows()
```

### Face detection by using the web camera

```python
webcam = cv. VideoCapture(0)
cam_status, img = webcam.read()
print("Frame captured successfully:", cam_status)
webcam.release()
```

```python
face_detect = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
def detect_faces():
    webcam = cv.VideoCapture(0)
    while True:
        _, frame = webcam.read()
        gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        faces = face_detect.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)
        for (x, y, w, h) in faces:
            cv.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv.imshow('Face Detection', frame)
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    webcam.release()
    cv.destroyAllWindows()
```

```python
detect_faces()
```
